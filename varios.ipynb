```python
import gradio as gr
from transformers import AutoTokenizer, AutoModelForCausalLM, AutoModelForSeq2SeqLM, pipeline

# Cargar modelos
bloom_model_name = "bigscience/bloom"
bloom_tokenizer = AutoTokenizer.from_pretrained(bloom_model_name)
bloom_model = AutoModelForCausalLM.from_pretrained(bloom_model_name)

translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-es")

mt5_model_name = "google/mt5-small"
mt5_tokenizer = AutoTokenizer.from_pretrained(mt5_model_name)
mt5_model = AutoModelForSeq2SeqLM.from_pretrained(mt5_model_name)

emotion_recognizer = pipeline("text-classification", model="bhadresh-savani/bert-base-go-emotion", return_all_scores=True)

# Funciones
def generate_bloom(prompt):
    inputs = bloom_tokenizer(prompt, return_tensors="pt")
    outputs = bloom_model.generate(inputs.input_ids, max_length=150)
    return bloom_tokenizer.decode(outputs[0], skip_special_tokens=True)

def translate_text(text, target_language="es"):
    return translator(text, max_length=500)[0]["translation_text"]

def generate_mt5(prompt):
    inputs = mt5_tokenizer(prompt, return_tensors="pt")
    outputs = mt5_model.generate(inputs.input_ids, max_length=150)
    return mt5_tokenizer.decode(outputs[0], skip_special_tokens=True)

def recognize_emotions(text):
    results = emotion_recognizer(text)
    emotions = {res["label"]: res["score"] for res in results[0]}
    sorted_emotions = sorted(emotions.items(), key=lambda x: x[1], reverse=True)
    return "\n".join([f"{label}: {score:.2f}" for label, score in sorted_emotions])

# Crear interfaz combinada
def combined_function(prompt, mode, language="es"):
    if mode == "Bloom Generation":
        return generate_bloom(prompt)
    elif mode == "Translation":
        return translate_text(prompt, language)
    elif mode == "mT5 Generation":
        return generate_mt5(prompt)
    elif mode == "Emotion Recognition":
        return recognize_emotions(prompt)

iface = gr.Interface(
    fn=combined_function,
    inputs=[
        gr.Textbox(label="Input Text"),
        gr.Dropdown(
            choices=["Bloom Generation", "Translation", "mT5 Generation", "Emotion Recognition"],
            label="Select Mode",
        ),
        gr.Textbox(label="Target Language (for Translation)", value="es", visible=False),
    ],
    outputs="text",
    title="Multi-Model AI with Emotion Recognition",
    description="Generate text with Bloom, translate with MarianMT, use mT5 for multilingual generation, or recognize emotions in text.",
)

if __name__ == "__main__":
    iface.launch()
```
